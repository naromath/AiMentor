파일을 입력 받아서 임베이딩해서 저장하는 것

1. 파일 로드
2. 파일에서 텍스트 추출
3. 텍스트 청크 나누기
4. embedding 셍성
5. 임베딩 저장을 위한 벡터 데이터 베이스 초기화
6. 파일에서 임베딩 생성
7. 벡터 데이터베이스 저장


```python
import os
import openai
from langchain_unstructured import UnstructuredLoader
from langchain_community.document_loaders import PyPDFLoader
import pdfplumber
```

```python
loader = PyPDFLoader("files/241227_2025년_소상공인_지원사업_통합_공고(최종).pdf")
data = loader.load()
```

```python
print(data)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
docs = text_splitter.split_documents(data)
```

```python
print(docs)
```

```python
def format_docs(docs):
    return "\n\n".join(document.page_content for document in docs)


print(format_docs(docs))
```

```python
from langchain_openai import OpenAIEmbeddings

embed = OpenAIEmbeddings(
    model="text-embedding-3-small",   
    api_key=os.environ["OPENAI_API_KEY"],

)
```

```python
for doc in docs:
    vector = await embed.aembed_query(doc.page_content)
    print(vector[:3])
```

```python
from langchain_community.vectorstores import FAISS

vector_store = FAISS.from_documents(docs, embed)
```

```python
vector_store.save_local("vector_store")
```

```python
retriever = vector_store.as_retriever()
```

```python
print(retriever)
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.5,

)
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """
            Answer the question using ONLY the following context. If you don't know the answer just say you don't know. DON'T make anything up.
            
            Context: {context}
            """,
        ),
        ("human", "{question}"),
    ]
)
```

```python
from langchain_core.runnables import RunnableLambda, RunnablePassthrough


chain = (
            {
                "context": retriever | RunnableLambda(format_docs),
                "question": RunnablePassthrough(),
            }
            | prompt
            | llm
)
```

```python
chain.invoke("소상공인 지원사업 종류는 무엇이 있나요?")
```

```python
chain.invoke("예비 창업자의 지금원은 얼마인가요?")
```
